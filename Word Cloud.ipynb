```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
import urllib
novel = urllib.request.urlopen('https://www.appsloveworld.com/wp-content/uploads/2018/10/Sample100.csv').read().decode("utf-8")
```

```python
stopwords = set(STOPWORDS)
novel_wc = WordCloud(background_color='White', max_words= 2000, stopwords = stopwords)
stopwords.add('pvt')
stopwords.add('ltd')
stopwords.add('publishing')
novel_wc.generate(novel)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
plt.imshow(novel_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np  # useful for many scientific computing in Python
from PIL import Image
```

```python
novel_mask = np.array(Image.open(urllib.request.urlopen('https://thumbs.dreamstime.com/z/big-ostrich-bird-white-background-have-path-181925505.jpg')))
```

```python
fig = plt.figure(figsize=(14, 18))

plt.imshow(novel_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
novel_wc = WordCloud(background_color='white', max_words=2000, mask=novel_mask, stopwords=stopwords)

# generate the word cloud
novel_wc.generate(novel)

# display the word cloud
fig = plt.figure(figsize=(14, 18))

plt.imshow(novel_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
